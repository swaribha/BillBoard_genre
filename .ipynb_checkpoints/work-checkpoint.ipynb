```python
import pandas as pd
from sqlalchemy import create_engine
from config import api_key 
import requests
import json
import pymongo
```

```python
csv_file = "billboard-1964-2015-songs-lyrics/billboard_lyrics_1964-2015.csv"
billboard_df = pd.read_csv(csv_file, encoding="latin-1")
billboard_df.head()
```

```python
billboard_df= billboard_df[['Rank', 'Song', 'Artist', 'Year']]
billboard_df.head()
```

```python

base_url = "http://ws.audioscrobbler.com/2.0/?method=track.getInfo"

playcount = []
tag1 = []
tag2 = []
tag3 = []
tag4 = []
tag5 = []

for index, row in billboard_df.iterrows():
    song = row['Song']
    artist = row['Artist']
    params = {
    "track": song,
    "artist": artist,
    "api_key": api_key,
    "format" : "json",
    "autocorrect[0|1]": 1
    }
    try:
        response = requests.get(base_url, params=params)
        lastfm_data = response.json()
        playcount.append(lastfm_data["track"]["playcount"])
        try:
            tag1.append(lastfm_data["track"]["toptags"]["tag"][0]["name"])
        except IndexError:
            tag1.append(0)
        try:
            tag2.append(lastfm_data["track"]["toptags"]["tag"][1]["name"])
        except IndexError:
            tag2.append(0)
        print(f"Processing record number {index+1}, {song}, {artist}!")
    except KeyError:
        playcount.append(0)
        tag1.append(0)
        tag2.append(0)
```

```python
billboard_df['Play Count'] = playcount
billboard_df['Tag 1'] = tag1
billboard_df['Tag 2'] = tag2
columns = ['Tag 1', 'Tag 2']
billboard_df
```

```python

conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.songs_db
db.song_info.drop()
records=billboard_df[:].to_dict(orient='records')
type(records)
db.song_info.insert_many(records)
```
